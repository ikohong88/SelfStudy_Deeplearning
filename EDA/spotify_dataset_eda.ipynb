# EDA (Exploratory Data Analysis) 탐색적 데이터 분석

```python
# -*- coding: utf-8 -*-
import pandas as pd

# read_csv 함수로 데이터를 Dataframe 형태로 불러옵니다.
# kaggle - spotify_dataset 파일 가져오기
file_path = 'data/spotify_dataset.csv'
spotify = pd.read_csv(file_path, sep = ',')
```

```python
print(spotify.shape)
print("------------------------------------")
print(spotify.info())
```

```python
# spotify 라는 Dataframe에서 순서대로 20개의 row 데이터를 보여줍니다.
spotify.head(20)
```

```python
spotify['Genre'] = spotify['Genre'].astype(str) # order_id는 숫자의 의미를 가지지 않기 때문에 str으로 변환합니다.
```

```python
print(spotify.describe()) # spotify dataframe에서 수치형 피처들의 요약 통계량을 확인합니다.
```

```python
print(len(spotify['Genre'].unique())) # Genre 개수를 출력합니다.
print(len(spotify['Highest Charting Position'].unique())) # Highest Charting Position의 개수를 출력합니다.
```

```python
# 가장 많이 플레이한 장르 Genre : top 20을 출력합니다.
Genre_count = spotify['Genre'].value_counts()[:20]
for idx, (val, cnt) in enumerate(Genre_count.iteritems(), 1):
    print("Top", idx, ":", val, cnt)
```
