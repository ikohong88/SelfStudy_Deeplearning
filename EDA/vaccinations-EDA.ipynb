# 1. 탐색

## 데이터셋의 기본 정보 확인

```python
# -*- coding: utf-8 -*-

import pandas as pd

# read_csv 함수로 데이터를 Dataframe 형태로 불러옵니다.
# 데이터셋 자료 : https://www.kaggle.com/umeshkumar017/vaccination-data
file_path = '../data/vaccinations.csv'
vaccine = pd.read_csv(file_path, sep = ',')
```

```python
print(vaccine.shape)
print("------------------------------------")
print(vaccine.info())
```

## 데이터셋의 행과 열 데이터

```python
# vaccine Dataframe에서 순서대로 10개의 row 데이터 출력
vaccine.head(10)
```

```python
print(vaccine.columns)
print("------------------------------------")
print(vaccine.index)
```

## 데이터셋의 수치적 특징 파악

```python
print(vaccine.describe()) # chipo dataframe에서 수치형 피처들의 요약 통계량을 확인합니다.
```

```python
# iso_code의 중복데이터를 제외한 개수를 출력합니다.
print(len(vaccine['iso_code'].unique()))
# date의 중복데이터를 제외한 개수를 출력합니다.
print(len(vaccine['date'].unique()))
```

# 2. 인사이트

```python
# 백신 데이터 수집 시작일 기준 location : top 10을 출력합니다.
location_vaccine_cnt = vaccine['location'].value_counts()[:10]
# enumerate : 인덱스와 원소로 이루워진 tuple을 만들어줌
# location_vaccine_cnt.iteritems() : location_vaccine_cnt의 (key,value) 쌍의 iterator를 return
for idx, (val, cnt) in enumerate(location_vaccine_cnt.iteritems(), 1):
    print("Top", idx, ":", val, cnt)
```

```python
vaccine['location'].value_counts().index.tolist()[30]
```

```python
# location당 백신 접종(날짜 기준) 횟수를 출력합니다.
vaccine_count = vaccine.groupby('location')['date'].count()
vaccine_count[:10] # item당 주문 개수를 출력합니다.
```

```python
# location당 하루 백신접종의 총량을 출력합니다.
vaccine_quantity = vaccine.groupby('location')['daily_vaccinations'].sum()
vaccine_quantity[:10] # location당 하루 백신접종의 총량을 출력합니다.
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 국가 이름 배열에 저장
vaccine_location_list = vaccine_quantity.index.tolist()
# 백신접종 국가 수
x_pos = np.arange(len(vaccine_location_list))
# 국가별 일일 백신접종 합계 데이터
vaccine_cnt = vaccine_quantity.values.tolist()
 
plt.bar(x_pos, vaccine_cnt, align='center')
plt.ylabel('location_vaccine_dailySum')
plt.title('Distribution of all total Vaccine')
 
plt.show()
```

## 한국 데이터 정보 EDA

```python
# 한국의 정보만 추출하기
is_vaccine_kor = vaccine['iso_code'] == 'KOR'
vaccine_kor = vaccine[is_vaccine_kor]
vaccine_kor.head(10)
```

```python
print(vaccine_kor.columns)
print("------------------------------------")
print(vaccine_kor.index)
```

```python
# vaccine_kor dataframe에서 수치형 피처들의 요약 통계량을 확인합
print(vaccine_kor.describe())
```

```python
# date의 중복데이터를 제외한 개수를 출력합니다.
print(len(vaccine_kor['date'].unique()))
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# 2차 접종 현황(예상 데이터)
kor_vaccine = vaccine_kor['daily_vaccinations']
# 1차 접종 현황(예상 데이터)
kor_vaccine2 = vaccine_kor['daily_vaccinations_raw']
x_pos = np.arange(len(kor_vaccine))
 
plt.bar(x_pos, kor_vaccine,color='lightblue', align='edge')
plt.bar(x_pos, kor_vaccine2, color='darkgreen',align='center')
plt.ylabel('daily_vaccinations_date')
plt.title('Distribution of KOR Vaccine')
 
plt.show()
```
